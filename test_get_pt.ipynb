```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
VIH = 1.8*0.7
VIL = 1.8*0.3
```

```python
CLK_data = np.load('./tmp/CH1_wave.npy')
DATA_data = np.load('./tmp/CH3_wave.npy')
DATA_rows  = [(idx, item) for idx,item in enumerate(DATA_data, start=1)]
CLK_rows  = [(idx, item) for idx,item in enumerate(CLK_data, start=1)]
```

```python
def get_DATA_pt(rows):
    tmp = []
    VIH_list = [] 
    for i in range(len(rows)-1):
        if rows[i][1] >= VIH:
            tmp.append(rows[i])

        if rows[i][1] <= VIL:
            tmp.append(rows[i])

    tmp_2 = []
    for i in range(len(tmp)):
        tmp_2.append(tmp[i])
        try:
            num = tmp[i][1]-(tmp[i+1][1])
            if abs(num) >=0.6:
                VIH_list.append(tmp_2)
                tmp_2 = []
        except Exception:
            VIH_list.append(tmp_2)
            tmp_2 = []
    return VIH_list
```

```python
DATA_Tf_pt = get_DATA_pt(DATA_rows)
CLK_Tf_pt = get_DATA_pt(CLK_rows)
```

```python
def plot_pt(DATA_rows, CLK_rows, DATA_Tf_pt, CLK_Tf_pt):
    DATA_PT_TMP = []
    CLK_PT_TMP = []

    for pt in DATA_Tf_pt:
        if pt[0][0] == 1 :
            #plt.scatter(pt[-1][0], pt[-1][1])
            DATA_PT_TMP.append([pt[-1][0], pt[-1][1]])
        elif pt[-1][0] == len(DATA_rows)-2 :
            #plt.scatter(pt[0][0], pt[0][1])
            DATA_PT_TMP.append([pt[0][0], pt[0][1]])
        else:
            #plt.scatter(pt[0][0], pt[0][1])
            DATA_PT_TMP.append([pt[0][0], pt[0][1]])
            #plt.scatter(pt[-1][0], pt[-1][1])
            DATA_PT_TMP.append([pt[-1][0], pt[-1][1]])

    for pt in CLK_Tf_pt:
        if pt[0][0] == 1 :
            #plt.scatter(pt[-1][0], pt[-1][1])
            CLK_PT_TMP.append([pt[-1][0], pt[-1][1]])
        elif pt[-1][0] == len(DATA_rows)-2 :
            #plt.scatter(pt[0][0], pt[0][1])
            CLK_PT_TMP.append([pt[0][0], pt[0][1]])
        else:
            #plt.scatter(pt[0][0], pt[0][1])
            CLK_PT_TMP.append([pt[0][0], pt[0][1]])
            #plt.scatter(pt[-1][0], pt[-1][1])
            CLK_PT_TMP.append([pt[-1][0], pt[-1][1]])

    return DATA_PT_TMP, CLK_PT_TMP
```

```python
DATA_PT_TMP, CLK_PT_TMP = plot_pt(DATA_rows, CLK_rows, DATA_Tf_pt, CLK_Tf_pt)
```
